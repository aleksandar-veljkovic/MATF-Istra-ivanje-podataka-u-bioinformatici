## Primer: Analiza diferencijalno eksprimiranih gena (DEG)
Analiza razlika u ekspresiji gena kod kontrolne grupe i osoba vakcinisanih trecom dozom

````python
data = pd.read_csv('GSE189263_tpm_of_25_samples.csv', delimiter='\t')
````

````python
# Izdvajanje uzoraka iz trazenih grupa i kolone sa nazivima gena
genes = data.iloc[:,[0]]
control = data.loc[:, ['Control-1', 'Control-2', 'Control-3', 'Control-4', 'Control-5']]
healthy_3rd = data.loc[:, ['Healthy-3rd-1', 'Healthy-3rd-2', 'Healthy-3rd-3', 'Healthy-3rd-4']]
````

````python
genes
````

````python
control
````

````python
healthy_3rd
````

````python
# Racunanje prosecnih vrednosti ekspresije gena kod kontrolne grupe
control_mean = pd.DataFrame(data=control.mean(axis=1), columns=['Control'])
control_mean
````

````python
# Racunanje prosecnih vrednosti ekspresije gena kod grupe osoba vakcinisanih trecom dozom
healthy_3rd_mean = pd.DataFrame(data=healthy_3rd.mean(axis=1), columns=['Healthy-3rd'])
healthy_3rd_mean
````

````python
# Racunanje log2FC - Log fold change, logaritma razlike u odnosima ekspresije izmedju grupa log2(B/A)
import numpy as np
c = 0.00001 # Korektivna konstanta za izbegavanje izracunavanja vrednosti log2(0)
log2fc_se = np.log2((healthy_3rd_mean.values + c) / (control_mean.values + c))
log2fc = pd.DataFrame(data=log2fc_se, columns=['log2FC'])
log2fc
````

````python
# Racunanje znacajnosti razlika izmedju srednjih vrednosti dve grupe uzoraka
from scipy.stats import ttest_ind

# Pairwise t-Test
p_values = pd.DataFrame(data = np.array([ttest_ind(control, healthy_3rd, axis=1).pvalue]).T, columns=['p'])
p_values
````

````python
merged_table = genes.join(log2fc).join(p_values).dropna()
merged_table
````

````python
# Vizualizacija podataka pomocu Volcano plot dijagrama
from bioinfokit.visuz import GeneExpression
GeneExpression.volcano(df=merged_table, sign_line=True, pv_thr=(0.05,0.05), lfc_thr=(10,10), lfc='log2FC', pv='p', geneid='Name', plotlegend=True)#, genenames='deg')
# Rezultujuci grafik se nalazi u fajlu volcano.png
````

````python
# Izdvajanje gena sa znajnom razlikom u ekspresiji
# Prag znacajnosti je p <= 0.05
sig_genes = merged_table[merged_table.p <= 0.05].sort_values(by='log2FC') # Sortirano prema log2FC
sig_genes
````

````python
down_regulated_genes = sig_genes[:100]
up_regulated_genes = sig_genes[-100:]
````

````python
down_regulated_genes
````

````python
up_regulated_genes
````

````python
# Vizualizacija razlika u ekspresiji gena pomocu heatmape
from seaborn import clustermap
````

````python
# Logaritmovane prosecne vrednosti TPM po grupama
log_mean_values = np.log(control_mean.join(healthy_3rd_mean)+c)
upregulated_means = up_regulated_genes.join(log_mean_values).iloc[:,-2:]
upregulated_means
clustermap(upregulated_means)
````
